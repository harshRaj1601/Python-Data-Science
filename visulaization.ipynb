# graphs

1. Univariate Graphs
    - Line Plot - for growth/decay over a period
    - Histogram - for distribution of Data
    - Density Plot - for distribution of data in Normalized form
    - Box Plot - for distribution of dara in quartiles
    - Pie Chart - for the distribution of Data in percentage
    - Bar Chart - for decrete data comparision

2. Bivariate Graphs
    - Scatter Plot - relation between 2 variables
    - Heat Map - relation between 2 variables in matrix form
    - Voilin Plot - relation between 2 variables in density form
    - Pair Plot - relation between 2 variables in grid form

3. Multivariate Graphs
    - Bubble Plot - relation between 2 variables with size
    - 3D Scatter Plot - relation between 3 variables on x, y, z axis
    - Stacked Bar Chart - relation between 3 variables in bar form

````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
````

````python
x = np.random.normal(size=1000)
plt.hist(x,bins=np.arange(-4,4,.5),edgecolor="black")
plt.show()
````

````python
sns.histplot(x,bins=np.arange(-4,4,0.5), kde=True)
````

````python
px.histogram(x, nbins=20)
````

````python
# box plot

x = np.random.normal(size=1000)
plt.boxplot(x)
plt.show()
````

````python
sns.boxplot(x)
````

````python
px.box(x)
````

````python
rainfall = [20,90,56,48,16,65]
mon = ["jan","feb","mar","apr","may","june"]
plt.pie(rainfall,radius=2,autopct="%.2f%%",labels=mon)
plt.show()
````

````python
px.pie(values=rainfall,names=mon,hole=.5)
````

````python
# area chart

x = np.arange(10)
y = np.random.randint(1,10,10)
plt.fill_between(x,y)
plt.show()
````

````python
px.area(x=x,y=y)
````

````python
years = np.arange(2010,2020,dtype=int)
sales = np.random.randint(1000,5000,10)
plt.bar(years,sales)
plt.show()
````

````python
sns.barplot(x=years,y=sales)
````

````python
px.bar(x=years,y=sales)
````

````python
px.bar_polar(r=years,theta=sales)
````

````python
x = np.random.normal(size=1000)
y = np.random.normal(size=1000)

plt.scatter(x,y,alpha=0.5,edgecolors="red")
````

````python
sns.scatterplot(x=x,y=y)
````

````python
px.scatter(x,y)
````

````python
z = np.random.normal(size=1000)
px.scatter_3d(x=x,y=y,z=z)
````

````python
sns.heatmap([[1,2],[3,4]],annot=True)
````

````python
sns.violinplot(x)
````

````python
sns.swarmplot(x)
````

````python
px.violin(x)
````

````python
z = np.random.randint(10,20,1000)
plt.scatter(x,y, s=z, alpha=.5,c=z)
````

````python
px.scatter(x=x,y=y,size=z, color=z)
````
